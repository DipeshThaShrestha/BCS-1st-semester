```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(df['age'], df['hours-per-week'], alpha=0.5)  # Create scatter plot
plt.title('Scatter Plot of Age vs. Hours per Week')  # Set title
plt.xlabel('Age')  # Set x-axis label
plt.ylabel('Hours per Week')  # Set y-axis label
plt.grid(True)  # Add gridlines
plt.show()  # Display the plot
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(df['capital-gain'], df['capital-loss'], alpha=0.5, label='Data Points')

# Set title and labels
plt.title('Scatter Plot of Capital Gain vs. Capital Loss')
plt.xlabel('Capital Gain')
plt.ylabel('Capital Loss')

# Add a legend
plt.legend()

# Add gridlines
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.scatter(df['age'], df['education-num'], alpha=0.5, label='Data Points')

# Set title and labels
plt.title('Scatter Plot of Age vs. Education Number')
plt.xlabel('Age')
plt.ylabel('Education Number')

# Add a legend
plt.legend()

# Add gridlines
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Sort the DataFrame by 'age'
df_sorted = df.sort_values(by=['age'])

# Create the line plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.plot(df_sorted['age'], df_sorted['hours-per-week'], label='Hours per Week')

# Set title and labels
plt.title('Trend of Hours per Week by Age')
plt.xlabel('Age')
plt.ylabel('Hours per Week')

# Add a legend
plt.legend()

# Add gridlines (optional)
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Group by 'age' and calculate the mean of 'capital-gain'
age_capital_gain = df.groupby('age')['capital-gain'].mean()

# Create the line plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.plot(age_capital_gain.index, age_capital_gain.values, marker='o', label='Average Capital Gain')

# Set title and labels
plt.title('Trend of Average Capital Gain by Age')
plt.xlabel('Age')
plt.ylabel('Average Capital Gain')

# Add a legend
plt.legend()

# Add gridlines (optional)
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Create a cross-tabulation of 'workclass' and 'education'
workclass_education = pd.crosstab(df['workclass'], df['education'])

# Create the clustered bar plot using seaborn
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
sb.countplot(x='workclass', hue='education', data=df)

# Set title and labels
plt.title('Distribution of Education Levels across Workclasses')
plt.xlabel('Workclass')
plt.ylabel('Count')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Add a legend
plt.legend(title='Education')

# Display the plot
plt.show()
```

```python
# Create a cross-tabulation of 'workclass' and 'education'
workclass_education = pd.crosstab(df['workclass'], df['education'])

# Create the stacked bar plot
workclass_education.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set title and labels
plt.title('Composition of Education Levels within Workclasses')
plt.xlabel('Workclass')
plt.ylabel('Count')

# Add a legend
plt.legend(title='Education Level')

# Display the plot
plt.show()
```

```python
# Assuming 'marital-status' is the group and 'relationship' is the category
# Create a cross-tabulation of 'marital-status' and 'relationship'
marital_relationship = pd.crosstab(df['marital-status'], df['relationship'])

# Create the stacked bar plot
marital_relationship.plot(kind='bar', stacked=True, figsize=(18, 8))

# Set title and labels
plt.title('Composition of Relationships within Marital Statuses')
plt.xlabel('Marital Status')
plt.ylabel('Count')

# Add a legend
plt.legend(title='Relationship')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=0,)

# Display the plot
plt.show()
```
